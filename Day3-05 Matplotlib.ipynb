```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,-4,10,5]

plt.plot(x,y)
plt.title("Sample Plot")
plt.xlabel("Days")
plt.ylabel("Celsius degrees")
plt.show()
```

```python
x = ["cat1", "cat2", "cat3", "cat4", "cat5"]
y = [2,3,-4,10,5]

plt.bar(x,y) # plt.plot(x,y), plt.bar(x,y)
plt.title("Sample Plot")
plt.xlabel("Days")
plt.ylabel("Celsius degrees", rotation=0)
plt.yticks(rotation=45)
plt.xticks(rotation=45)
plt.show()
```
