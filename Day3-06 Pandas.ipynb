```python
pivot_table = pd.pivot_table(df2, values='total (R$)', index='rooms', columns='city', aggfunc='mean')
pivot_table
```

```python
import matplotlib.pyplot as plt
pivot_table.plot(kind='bar', figsize=(10, 6))
plt.title('Average Cost by City and Number of Rooms')
plt.xlabel('City')
plt.ylabel('Average Cost')
plt.legend(title='Rooms')
plt.show()
```
